```python
import os

#'http://tisvcloud.freeway.gov.tw/history/TDCS/M06A/20191119/23/TDCS_M06A_20191119_230000.csv'
TDCS_M06A = "https://tisvcloud.freeway.gov.tw/history/TDCS/M06A/"

Year = "2023"
Month = "10"
Day = "05"
Hour_Start = 0
Hour_End =23


OutputDirTag ="jdwang"

OutPutDir = "./TDCS_Download_"+OutputDirTag+"/M06A/"+Year+Month

#os.mkdir(OutPutDir)
if not (os.path.exists(OutPutDir)):
    os.makedirs(OutPutDir)
    print("Dir Created "+OutPutDir)
else:
    print("Dir Existed "+OutPutDir)
```

```python
import requests
import pandas as pd
import os
import time

# TDCS_M06A = "https://tisvcloud.freeway.gov.tw/history/TDCS/M06A/"

for Hour in range(Hour_Start, Hour_End+1):
    if Hour < 10:
        HourStr = "0"+str(Hour)
        print("0"+str(Hour))
    else:
        HourStr = str(Hour)
        print (str(Hour))

    OneDate = Year+Month+Day

    OneDate_OneHour_FileName = "TDCS_M06A_"+OneDate+"_"+HourStr+"0000.csv"

    url = TDCS_M06A +OneDate+'/'+HourStr+'/'+ OneDate_OneHour_FileName

    #csv_data = pd.read_csv(url,encoding='utf-8', header=None)
    csv_data = pd.read_csv(url,encoding='utf-8', header=None, names=FieldNames)

    OutputFileLocation = OutPutDir+"/"+OneDate_OneHour_FileName

    csv_data.to_csv(OutputFileLocation)
    print("Download=>"+OutputFileLocation)
    time.sleep(30)
```
